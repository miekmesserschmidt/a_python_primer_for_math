Basic plotting with matplotlib
=========

Basic line plots
----------------

```python
# To make plots we must import 
# matplotlib.pyplot and numpy
import matplotlib.pyplot as plt
import numpy as np
```

```python
# We create a figure and axes do make our plot on
figure, axes = plt.figure(), plt.axes()

# We take 10 evenly spaced x-values in the interval (-1,1)
X = np.linspace(-1, 1, 10)
# We compute the Y-values
Y = X**2 + 2

# We plot the data on the axes
axes.plot(X, Y)

# We show the plot
plt.show()
```

```python
# Let's plot the function x |-> sin(x) + .5*sin(4x)  with in x in (-pi,3pi)
figure, axes = plt.figure(), plt.axes()

X = np.linspace(-np.pi, 3*np.pi, 100)
Y = np.sin(X) + .5*np.sin(4*X)

axes.plot(X, Y)
plt.show()
```

Basic scatter plots
-----------------

```python
# We import matplotlib.pyplot and numpy
import matplotlib.pyplot as plt
import numpy as np
```

```python
# We plot a list of points on a set of axes
# See if you can match each dot to its coordinates
figure, axes = plt.figure(), plt.axes()

points = [
    (4  , 5),
    (5  , 5),
    (5.5, 5.5),
    (3  , 3.45),
    (3.6, 4.36),
]

xvals, yvals = zip(*points)

axes.plot(xvals, yvals, "o")
plt.show()
```

Parametric plots
----------------

```python
import matplotlib.pyplot as plt
import numpy as np
# Let's plot the parametric plot
# t |-> (cos(t), sin(t)) for t in (0,4)

figure, axes = plt.figure(), plt.axes()

# We take 100 evenly spaced t-values in the interval (0,4)
T = np.linspace(0,4, 100)
# ... and compute the x- and y-values
X = np.cos(T)
Y = np.sin(T)

plt.plot(X, Y)
plt.show()
```

Changing the aspect ratio, plot range and size
----------------

```python
# The following plot is a bit squashed.
# That is because the "aspect ratio" is not equal to 1
# See: https://en.wikipedia.org/wiki/Aspect_ratio_(image)
import matplotlib.pyplot as plt
import numpy as np
figure, axes = plt.figure(), plt.axes()

T = np.linspace(0,4, 100)
X = np.cos(T)
Y = np.sin(T)

plt.plot(X, Y)
plt.show()
```

```python
# ... we can make this better by setting the 
# axes' "aspect ratio"  equal to 1

import matplotlib.pyplot as plt
import numpy as np

figure, axes = plt.figure(), plt.axes()
axes.set_aspect(1)

T = np.linspace(0,4, 100)
X = np.cos(T)
Y = np.sin(T)

plt.plot(X, Y)
plt.show()
```

```python
# We can restrict our plot ranges over 
# which the plot is displayed
# using "set_xlim" and "set_ylim" on the axes.

figure, axes = plt.figure(), plt.axes()
axes.set_aspect(1)

# We restrict to the first quadrant
axes.set_xlim(0,1)
axes.set_ylim(0,1)

T = np.linspace(0,4, 100)
X = np.cos(T)
Y = np.sin(T)

axes.plot(X, Y)
plt.show()

```

```python
# We can nmake our plot a bit larger
figure  = plt.figure(figsize=(10,10)) # in inches!   D:<  
axes = plt.axes()
axes.set_aspect(1)


T = np.linspace(0,4, 100)
X = np.cos(T)
Y = np.sin(T)

axes.plot(X, Y)


plt.show(figure)
```

Plotting with sympy
-------------------

```python
# Let's plot the cubic x**3 -x**2 +x +3 with x in (-1,1)
# using sympy

# We will need the symbol "x"
from sympy.abc import x
import sympy

# We define the figure and axes.
figure, axes = plt.figure(), plt.axes()

# We define the function f(x) := x**3 - x**2 + x +3
f = sympy.lambdify([x], x**3 - x**2 + x +3)


X = np.linspace(-1, 1, 100)
Y = f(X)

axes.plot(X, Y)
plt.show()
```

Multiple plots on the same axis
--------------------------

```python
# Let's plot two functions on the interval (-2,2)
# using sympy. 

# We will need the symbol "x"
from sympy.abc import x
import sympy

figure, axes = plt.figure(), plt.axes()

f = sympy.lambdify([x], x**3 - x**2 + x +3)
g = sympy.lambdify([x], -x**2 + x +2)

X = np.linspace(-2,2,100)

axes.plot(X, f(X))
axes.plot(X, g(X))
plt.show()
```
